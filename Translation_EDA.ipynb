```python
import nltk
nltk.download('punkt')
from nltk import sent_tokenize
import pandas as pd
import re
```

```python
svspArticle = """
Lördagens omgång på Stryktipset slutade med miljonutdelning för andra helgen i rad. Det var återigen i Norge där det skrällde som värst. Miljonvinsterna gick till Norrköping och Sollentuna. I Sollentuna löste en man 13 rätt på 128 kronor satsade och han tackar Teodor Haltvik i KFUM Oslo för miljonvinsten.

– Fantastiskt, men allt är som i en bubbla. Jag har inte pratat med en människa om min vinst. Min fru är bortrest, jag vill berätta för henne först och inte via telefon. Så situationen är helt låst, berättar miljonvinnaren.

Det var på söndagsmorgonen som mannen såg ett mail från Svenska Spel Sport & Casino med ordet högvinst.

– Det var så raffinerat, ordet högvinst. Men istället för att logga in och se vinstsumman, tog jag en paus och gjorde kaffe. Samtidigt malde tankarna på hur mycket är en högvinst. När jag sen såg vinstsumman kom tårarna, fram kom vår lilla hund och i ett försök att slicka bort tårarna som hon inte förstod var av glädje.

Vinsten var på 1 492 378 kronor. Starkast bidragande till stora vinstsumman var KFUM Oslo som slog Molde. KFUM Oslo var streckat till endast 8 procent.

– Jag är så glad att det var på just Stryktipset jag vann. För nu har jag kunnat gå tillbaka och se matcherna. Molde mot KFUM Oslo var lika när Teodor Haltvik fick bollen. Jag vill tacka Teodor för han inte sköt. Han var fri och kunde skjuta men han passade i stället och KFUM Oslo kunde göra ett säkrare mål, avslutar miljonvinnaren från Sollentuna.
"""
```

```python
df = pd.DataFrame({
    "Language": ["Original Article"],
    "Article": re.sub(r'\s+', ' ', svspArticle).strip()
})
```

## Llama2

```python
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
```

```python
model = MBartForConditionalGeneration.from_pretrained("SnypzZz/Llama2-13b-Language-translate")
tokenizer = MBart50TokenizerFast.from_pretrained("SnypzZz/Llama2-13b-Language-translate", src_lang="sv_SE")
```

```python
sentences = sent_tokenize(svspArticle)

translated_sentences = []
```

```python
for sentence in sentences:
    model_inputs = tokenizer(sentence, return_tensors="pt", truncation=True)

    generated_tokens = model.generate(
        **model_inputs,
        forced_bos_token_id=tokenizer.lang_code_to_id["en_XX"]
    )

    translated_sentence = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)[0]
    translated_sentences.append(translated_sentence)

translated_paragraph = " ".join(translated_sentences)

print(translated_paragraph)
```

## Nordic Sentence completion

```python
import torch
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM

# Initialize Variables
model_name = "AI-Sweden-Models/gpt-sw3-126m-instruct"
device = "cuda:0" if torch.cuda.is_available() else "cpu"
prompt = "Träd är fina för att"

# Initialize Tokenizer & Model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
model.eval()
model.to(device)
```

```python
input_ids = tokenizer(prompt, return_tensors="pt")["input_ids"].to(device)

generated_token_ids = model.generate(
    inputs=input_ids,
    max_new_tokens=100,
    do_sample=True,
    temperature=0.6,
    top_p=1,
)[0]

generated_text = tokenizer.decode(generated_token_ids)
```

## facebook/mbart-large-50-many-to-many-mmt

```python
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
```

```python
model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-many-to-many-mmt", src_lang="sv_SE")
```

```python
model_inputs = tokenizer(svspArticle, return_tensors="pt")

generated_tokens = model.generate(
    **model_inputs,
    forced_bos_token_id=tokenizer.lang_code_to_id["en_XX"]
)

translated_article = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)

print(translated_article)
```

```python
temp_df_row = pd.DataFrame({
    "Language": ["Translated using Bart Model"],
    "Article": re.sub(r'\s+', ' ', translated_article[0]).strip()
})

df = pd.concat([df, temp_df_row], ignore_index=True)
```

### Truncate then translate

```python
sentences = sent_tokenize(svspArticle)
```

```python
translated_sentences = []
for sentence in sentences:
    model_inputs = tokenizer(sentence, return_tensors="pt", truncation=True)

    generated_tokens = model.generate(
        **model_inputs,
        forced_bos_token_id=tokenizer.lang_code_to_id["en_XX"]
    )

    translated_sentence = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)[0]
    translated_sentences.append(translated_sentence)

translated_article = " ".join(translated_sentences)

print(translated_article)
```

```python
temp_df_row = pd.DataFrame({
    "Language": ["Translated using Bart model with truncation"],
    "Article": re.sub(r'\s+', ' ', translated_article).strip()
})

df = pd.concat([df, temp_df_row], ignore_index=True)
```

## Marian Model

```python
from transformers import MarianMTModel, MarianTokenizer

model_name = "Helsinki-NLP/opus-mt-sv-en"
model = MarianMTModel.from_pretrained(model_name)
tokenizer = MarianTokenizer.from_pretrained(model_name)

model_inputs = tokenizer(svspArticle, return_tensors="pt")

translated_tokens = model.generate(**model_inputs)

translated_article = tokenizer.decode(translated_tokens[0], skip_special_tokens=True)

print(translated_article)
```

```python
temp_df_row = pd.DataFrame({
    "Language": ["Translated using Marian model"],
    "Article": re.sub(r'\s+', ' ', translated_article).strip()
})

df = pd.concat([df, temp_df_row], ignore_index=True)
```

## Test to Speech with microsoft

```python
from transformers import pipeline
from datasets import load_dataset
import soundfile as sf
```

```python
import torch
```

```python
synthesiser = pipeline("text-to-speech", "microsoft/speecht5_tts")

embeddings_dataset = load_dataset("Matthijs/cmu-arctic-xvectors", split="validation")
```

```python
speaker_embedding = torch.tensor(embeddings_dataset[1306]["xvector"]).unsqueeze(0)


speech = synthesiser(
    "Fantastiskt, men allt är som i en bubbla. Jag har inte pratat med en människa om min vinst",
    forward_params={"speaker_embeddings": speaker_embedding}
)

sf.write("speech1.wav", speech["audio"], samplerate=speech["sampling_rate"])
```
